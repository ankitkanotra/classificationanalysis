```python
#Handling outlier through IQR
```

```python
IQR = dataset['Age'].quantile(0.75) - dataset['Age'].quantile(0.25)
```

```python
IQR
```

```python
lower_bound=dataset['Age'].quantile(0.75)-1.5*IQR
upper_bound=dataset['Age'].quantile(0.75)+1.5*IQR
print(upper_bound)
```

```python
dataset.loc[dataset['Age']>upper_bound,'Age']=upper_bound
dataset.loc[dataset['Age']<lower_bound,'Age']=lower_bound
```

```python
IQR_Score = dataset['Score'].quantile(0.75) - dataset['Score'].quantile(0.25)
```

```python
lower_bound_Score=dataset['Score'].quantile(0.75)-1.5*IQR_Score
upper_bound_Score=dataset['Score'].quantile(0.75)+1.5*IQR_Score
print(upper_bound_Score)
```

```python
dataset.loc[dataset['Score']>upper_bound_Score,'Score']=upper_bound_Score
dataset.loc[dataset['Score']<lower_bound_Score,'Score']= lower_bound_Score
```

```python
X = dataset.loc[:,['Score', 'Nationality', 'Age', 'Tenure', 'Balance', 'Products', 'Card',
       'Active', 'Salary','Male']]
```

```python
y=dataset.iloc[:,-2]
```

```python
#import train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
standard_scaler = preprocessing.StandardScaler()
```

```python
pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
xgb = XGBClassifier()
xgb.fit(X_train,y_train)

y_xgb = xgb.predict(X_test)

cmx = confusion_matrix(y_test,y_xgb)
true_01,false_01 = cmx[0][0],cmx[0][1]
true_02,false_02 = cmx[1][0],cmx[1][1]
total = cmx[0][0] + cmx[0][1] + cmx[1][0] + cmx[1][1]
print(f'T predictions: {true_01+true_02}\nF predictions: {false_01+false_02}')
print(f'Accuracy ratio from XGBoost: %{round(((true_01+true_02)/(total))*100,5)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```
